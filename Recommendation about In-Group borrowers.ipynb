## Investigation Overview

> Investigating if in group borrowers deserve thier recomendations.

## Dataset Overview

> Prosper loan dataset. It contains data of about 113937 loans and 81 features for these lonas.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
loans = pd.read_csv('prosperLoanData.csv')
#wrangling dataframe
loans = loans.replace({'LoanStatus' : { 'Past Due (1-15 days)' : 'Past Due' , 'Past Due (31-60 days)' : 'Past Due' , 'Past Due (61-90 days)' : 'Past Due' , 'Past Due (91-120 days)' : 'Past Due' , 'Past Due (16-30 days)' : 'Past Due' , 'Past Due (>120 days)' : 'Past Due'}})
loans = loans.replace({'LoanStatus' : { 'Current' : 'Undue' , 'Completed' : 'Undue'}})
loans = loans[(loans['LoanStatus'] != 'Cancelled') & (loans['LoanStatus'] != 'FinalPaymentInProgress')]
['In Group', 'Out of Group'] ,
loans = loans.replace({'CurrentlyInGroup' : { True : 'In Group' , False : 'Out of Group'}})
```

## Do in group borrowers have higher recomendations?

> Yes ,in group borrowers have higher recomendations in average than out of group borrowers

```python


plt.figure(figsize = [12,8])
base_color = sb.color_palette()[0]
sb.barplot( data = loans , x = 'CurrentlyInGroup' , y = 'Recommendations' , color=base_color);
plt.title('Borrowers Average Recommendations Based on Group Joining ' , fontsize=20)
plt.xlabel('Group Joining', fontsize=20);
plt.ylabel('Avg Recommendations', fontsize=20);
plt.xticks(fontsize=15);
```

## They also have higher score from Prosper

```python
plt.figure(figsize = [12,8])
base_color = sb.color_palette()[0]
sb.boxplot( data = loans , x = 'CurrentlyInGroup' , y = 'ProsperScore' , color=base_color);
plt.title('Borrowers Average Prosper Score Based on Group Joining ', fontsize=20)
plt.xlabel('Group Joining', fontsize=20);
plt.ylabel('Avg Prosper Score', fontsize=20);
plt.xticks(fontsize=15);
```

## What about delinquent days?

> Despite in group borrowers have higher recomendations, they have higher delinquent days..

```python
plt.figure(figsize = [12,8])
base_color = sb.color_palette()[0]
sb.barplot( data = loans , x = 'CurrentlyInGroup' , y = 'LoanCurrentDaysDelinquent' , color = base_color);

plt.title('Borrowers Average delinquent days Based on Group Joining ' , fontsize=20)
plt.xlabel('Group Joining' , fontsize=20);
plt.ylabel('Avg Delinquent Days' , fontsize=20);
plt.xticks(fontsize=15);
```

# What about principal loss?
> Lenders incure more principal loss with in group borrowers

```python
plt.figure(figsize = [12,8])
base_color = sb.color_palette()[0]
sb.barplot( data = loans , x = 'CurrentlyInGroup' , y = 'LP_GrossPrincipalLoss' , color=base_color);

plt.title('Borrowers Average Principal Loss Based on Group Joining ' , fontsize=20)
plt.xlabel('Group Joining' , fontsize=20);
plt.ylabel('Avg Principal Loss' , fontsize=20);
plt.xticks(fontsize=15);
```

## Is this the case in all loan status?
> In Past Due and Defualted loans there is a small difference but the problem is clear in charged off loans.


```python

plt.figure(figsize = [14,8])
sb.barplot( data = loans , x = 'LoanStatus' , y = 'LoanCurrentDaysDelinquent' , hue = 'CurrentlyInGroup')
plt.title('In group borrowers  have higher delinquent days just in charged off loans ' , fontsize=20)
plt.xlabel('Loan Status' , fontsize=20)
plt.ylabel('Avg. Delinquent Days' ,  fontsize=20)
plt.legend( fontsize = 'xx-large' , ncol = 2 );
plt.xticks(fontsize= 15);
```
